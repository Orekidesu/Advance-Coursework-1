# Create new Data frame
### Feature Selection
    
   - in this project, I'll only be using 4 features:
       - id
       - title
       - genre
       - overview
   - for the 'genre' and 'overview', i'll be merging it later...

```python
movies = movies[['id','title','genre','overview']]
movies
```

# Merging the 'genre' and 'overview' feature and create new column

- this is important since I'm only be using content base filtering method/technique in my movie recommendation system 

```python


# create new column called 'tags' containing the data in the 'overview' & 'genre'
movies['tags'] = movies['genre']+movies['overview']
movies
```

# Delete/Drop the 'overview' and 'genre' in our new dataset
- This is to avoid any redundancy in our data frame
    


```python
new_data = movies.drop(columns = ['overview', 'genre'])
new_data
```

# Applying the algorithm
- we need to convert our text into a vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
#Output will be 10000x10000 matrix
```

# Using the Cosine Similarity Technique

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

### Showing the Similarity Value of each row

```python
similarity
```

# Accessing the Index no. and getting the distance using the similarity

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
distance
```

# Accessing the first 5 closest similarity

```python
#Except from the first index get the first 5 closest similarity
for i in distance[1:6]:
    print(new_data.iloc[i[0]].title)
```

```python
#create a function to test if the algorithm works
def recommend(movies):
    index = new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in distance[1:6]:
        print(new_data.iloc[i[0]].title)
```

```python
#pass some data in the function
recommend("Barbie of Swan Lake")
```
